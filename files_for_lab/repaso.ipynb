Learning how to use map( ), lambda and reduce( )

```python
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]

result = map(lambda x, y: x + y, numbers1, numbers2)
print(list(result))
```

```python
radio = [1.5,2,3.2,4,6,7.7]
result = map(lambda r: math.pi * r**2, radio)
print(list(result))
```

```python
def area(r):
    return math.pi * r**2
radio = [1.5,2,3.2,4,6,7.7]

resultado = (map(area, radio))
print(list(resultado))
```

```python
# We are getting rid of % symbol and convert the column from object type to numeric (float64)
df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace("%","")
df['customer_lifetime_value'] = df['customer_lifetime_value'].astype(np.number)
```

```python
def porcien(n):
    return 100 * n

df['customer_lifetime_value'] = list(map(porcien, list(df['customer_lifetime_value'])))
df['customer_lifetime_value']
```
