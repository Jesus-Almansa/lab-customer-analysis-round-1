### Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

```python
df.columns
```

```python
# We remove the & symbol
df["customer lifetime value"] = df["customer lifetime value"].str.replace('\%', '', regex=True)
# We transform the data from object to numeric in order to operate
df['customer lifetime value'] =  pd.to_numeric(df['customer lifetime value'], errors='coerce')

df['customer lifetime value'] = list((map(lambda x: np.dot(x,1/100), df['customer lifetime value'])))
df
```
